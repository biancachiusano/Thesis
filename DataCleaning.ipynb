```python
overall_data = pd.DataFrame(columns=['Case', 'Violation'])
```

```python
filenames = []
directories = []
for filename in os.listdir('all_ten'):
    filenames.append(filename)
    file_path = os.path.join('all_ten', filename)
    directories.append(file_path)
```

```python
# Some files seem to not have the discussion section
files_to_consider = []
#count = 0
for file in directories:
    with open(file) as f:
        lines = f.readlines()
    filter_object = list(filter(lambda a: 'FIRST SECTION' in a, lines))
    if len(filter_object) == 0:
        filter_object = list(filter(lambda a: 'SECOND SECTION' in a, lines))
    if len(filter_object) == 0:
        filter_object = list(filter(lambda a: 'THIRD SECTION' in a, lines))
    if len(filter_object) == 0:
        filter_object = list(filter(lambda a: 'FOURTH SECTION' in a, lines))
    if len(filter_object) == 0:
        filter_object = list(filter(lambda a: 'FIFTH SECTION' in a, lines))
    if len(filter_object) != 0:
        files_to_consider.append(file)
```

```python
def clean(file):
    with open(file) as f:
        lines = f.readlines()
    filter_object = list(filter(lambda a: 'FIRST SECTION' in a, lines))
    if len(filter_object) == 0:
        filter_object = list(filter(lambda a: 'SECOND SECTION' in a, lines))
    if len(filter_object) == 0:
        filter_object = list(filter(lambda a: 'THIRD SECTION' in a, lines))
    if len(filter_object) == 0:
        filter_object = list(filter(lambda a: 'FOURTH SECTION' in a, lines))
    if len(filter_object) == 0:
        filter_object = list(filter(lambda a: 'FIFTH SECTION' in a, lines))
    cleaned = filter_object[0].replace('\xa0', ' ')
    return cleaned
```

## QUESTION: is it always holds that..?

IF THE LABEL_VIOLATION IS: "no violation of Article 10" THE VALUE COUNTS IS:

```python
def label_violation(case):
    no_violation = "Holds that there has been no violation of Article 10"

    if no_violation in case:
        return 0
    else:
        return 1
```

```python
for file in files_to_consider:
    #print(file)
    cleaned = clean(file)
    filename = file.replace('.txt', '').replace('all_ten/', '')
    new_row = {'Case': filename, 'Violation': label_violation(cleaned)}
    overall_data = overall_data.append(new_row, ignore_index = True)
```
